**so the problwm was that the accident coulmn of my actual dataset contained only 1 value that was "NO" which was not suitable for data
because for training it must have more than one class of values so i changed 25% of the features values to "YES" by using the code below**

```python
import pandas as pd
import numpy as np



# Calculate the number of rows to set as "Yes" in the "Accident" column
total_rows = len(df)
desired_percentage = 25
num_rows_to_set_yes = int(total_rows * (desired_percentage / 100))

# Shuffle the DataFrame rows to randomize the selection
df_shuffled = df.sample(frac=1, random_state=42)

# Set the first 'num_rows_to_set_yes' rows in the 'Accident' column as "Yes"
df_shuffled.loc[:num_rows_to_set_yes, 'Accident'] = 'Yes'

# Set the rest of the rows in the 'Accident' column as "No"
df_shuffled.loc[num_rows_to_set_yes:, 'Accident'] = 'No'

# Assign the modified "Accident" column back to the original DataFrame
df['Accident'] = df_shuffled['Accident']
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt


# Select features and target variable
X = df[['Gender', 'Violation Type']]
y = df['Accident']


X = pd.get_dummies(X, drop_first=True)

#  i am Splitting  the dataset into training and testing  the training size is 80% of actual data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)


model = LogisticRegression()
model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = model.predict(X_test)



# Visualize the number of violations caused by each gender
gender_violations = df.groupby('Gender')['Violation Type'].value_counts().unstack()
gender_violations.plot(kind='bar', stacked=True)
plt.xlabel('Gender')
plt.ylabel('Number of Violations')
plt.title('Number of Violations by Gender')
plt.legend(title='Violation', loc='upper right', labels=['Citation', 'Warning'])
plt.show()
```
